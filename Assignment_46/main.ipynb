```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from perceptron import Perceptron
```

# 1-Employee's salary ðŸ’µ

- Make a linear dataset 

```python
x = np.random.normal(10,1,100)
y = x + np.random.normal(10,0.5,100)

plt.scatter(x,y)
plt.show()
```

- Split data to train and test datasets

```python
x_train , x_test , y_train , y_test = train_test_split(x , y,shuffle=True)
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

- Fit your model on the employee's salary dataset

```python
df = Perceptron(4,5)
w,b=df.fit(x_train , y_train)
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,x_train*w+b)
```

```python
plt.scatter(x_test,y_test,color="blue")
a=df.predict(x_test)
plt.plot(x_test,a)
```

```python
df.evaluate(x_test , y_test)
```
